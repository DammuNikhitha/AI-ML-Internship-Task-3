# Task 3 : Exploratory Data Analysis(EDA)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["species"] = iris.target
df["species"] = df["species"].map({0: "setosa", 1: "versicolor", 2: "virginica"})

df.head()
```

## Dataset Overview
- The Iris dataset contains measurements of iris flowers.
- Numerical features include sepal and petal dimensions.
- Target variable is flower species.

```python
df.info()
```

# Step 1 : Distribution of Numerical Features(Histograms)
## Distribution of Numerical Features
Histograms help understand the spread and frequency of numerical data.

```python
df.hist(figsize=(10, 8))
plt.tight_layout()
plt.show()
```

** Insight:**

Petal length and petal width show clear separation among species.
Sepal features overlap more

# Step 2 : Categorical Feature Analysis(Count Plot)

Count plots show the distribution of categorical variables.

```python
sns.countplot(x="species", data=df)
plt.show()
```

** Insight:**

All three species have equal number of samples.
Dataset is balanced.

# Step 3 : Box Plots for Outlier Detection
Box plots help identify extreme values and spread.

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df.drop("species", axis=1))
plt.show()
```

** Insight:**

Some outliers are present in sepal width.
Petal features have cleaner distributions.

# Step 4 : Correlation Heatmap
Correlation measures the strength of relationship between features.

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.drop("species", axis=1).corr(), annot=True, cmap="coolwarm")
plt.show()
```

** Insight:**

Petal length and petal width are highly correlated.
Sepal features have weaker correlations.

# Step 6: Important Features for Prediction
## Important Features
- Petal length
- Petal width

These features show strong correlation and good class separation,
making them important for prediction.

# Step 7 : Summary of Findings
## Summary of EDA Findings
- Dataset is clean and balanced
- Petal features are most informative
- Some minor outliers exist
- Strong correlation between petal length and petal width
- Visualization makes patterns easy to understand
